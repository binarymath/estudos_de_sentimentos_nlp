```python
# Importando a pipeline, da biblioteca transformers

from transformers import pipeline
```

```python
classifier = pipeline('sentiment-analysis')
```

```python
classifier("Isto não é muito bacana")
```

```python
classifier("Parabéns pelo trabalho !")
```

```python
classifier("Sou fã da minha família")
```

```python
def score(review_text):
  return classifier(review_text)[0]['label']
```

```python
test_df = reviews['test'].to_pandas()
```

```python
test_df['model_prediction'] = test_df['text'].apply(score)
```
